```python
#Newton Method Function 1
from scipy import misc
import numpy as np
import timeit
from scipy.misc import derivative
def f(x):
    return np.exp(-2*x) -x - np.sin(x) +2
def g(x):
    return derivative(f, x, dx = 1e-6)

def newtonRaphson(x0,e,N):
    print('\n\n*** NEWTON Method ***')
    step = 1
    flag = 1
    condition = True
    st1 = timeit.default_timer()
    while condition:
        if g(x0) == 0.0:
            print('0 in down.')
            break
        x1 = x0 - f(x0)/g(x0)
        print('Step: %d, x1 = %0.6f ve f(x1) = %0.6f' % (step, x1, f(x1)))
        x0 = x1
        step = step + 1
        if step > N:
            flag = 0
            break
        condition = abs(f(x1)) > e
    st2 = timeit.default_timer()
    if flag==1:
        print('\nFinding root: %0.8f' % x1)
    else:
        print('\nDoes not converges.')

x0 = 1
e = 10**(-5)
N = 20
newtonRaphson(x0,e,N)
print("Work time: {0}ms\n".format((st2-st1)*1000))
```
